## FM FORECAST PROD & INJ (XLSX MODE)

```python
import pandas as pd
import xlwings as xw

df_prod = pd.read_excel("D:\GG\MINAS\SENSITIVITY LIQUID & INJECTION MFFM - PHR REQUEST AMEL\Liquid and Injection HM149_Sensitivity.xlsx", 
sheet_name = "Liquid Sensitivity (no 0)", skiprows=1)
df_inj = pd.read_excel("D:\GG\MINAS\SENSITIVITY LIQUID & INJECTION MFFM - PHR REQUEST AMEL\Liquid and Injection HM149_Sensitivity.xlsx", 
sheet_name = "Injection Sensitivity (no 0)", skiprows=1)

wb = xw.Book("D:\GG\MINAS\SENSITIVITY LIQUID & INJECTION MFFM - PHR REQUEST AMEL\Coding GG\Liquid FM.xlsx")

for sheet in wb.sheet_names:

    ws = wb.sheets[sheet]

    rowcount = 0

    ws.range(rowcount + 1, 1).value = "MODEL_DEFINITION"
    ws.range(rowcount + 3, 1).value = "Forecast Rate " + str(sheet)
    ws.range(rowcount + 5, 1).value = "START"
    ws.range(rowcount + 7, 1).value = "DATE \"1-JAN-2024\" "
    ws.range(rowcount + 9, 1).value = "HistoricalDataControl \"New Historical Data Control 1\" {"
    ws.range(rowcount + 10, 1).value = "   Wells=[]"
    ws.range(rowcount + 11, 1).value = "}"

    if "Prod" in sheet:

        #producer
        for well in df_prod.index:

            ws.range(rowcount + 13, 1).value = "Well " + "\"" + str(df_prod['Wells'][well]) + "\"" + "   {"
            ws.range(rowcount + 14, 1).value =  "     " + "Type=PRODUCER"
            
            if int(df_prod[sheet][well]) != 0:
                ws.range(rowcount + 15, 1).value = "     " + "Status=OPEN"
            else:
                ws.range(rowcount + 15, 1).value = "     " + "Status=ALL_COMPLETIONS_SHUTIN"

            ws.range(rowcount + 16, 1).value = "     " + "Constraints=[Constraint(" + str(int(df_prod[sheet][well])) + " LIQUID_PRODUCTION_RATE)]"
            ws.range(rowcount + 17, 1).value = "     " + "HistoricalControlModes=UNSET"
            ws.range(rowcount + 18, 1).value = "     " + "HistoricalData=UNSET"
            ws.range(rowcount + 19, 1).value = "}"

            rowcount += 7
    else:

        #injector
        for well in df_inj.index:

            ws.range(rowcount + 13, 1).value = "Well " + "\"" + str(df_inj['Wells'][well]) + "\"" + "   {"
            ws.range(rowcount + 14, 1).value = "     " + "Type=INJECTOR"

            if int(df_inj[sheet][well]) != 0:
                ws.range(rowcount + 15, 1).value = "     " + "Status=OPEN"
            else:
                ws.range(rowcount + 15, 1).value = "     " + "Status=ALL_COMPLETIONS_SHUTIN"

            ws.range(rowcount + 16, 1).value = "     " + "Constraints=[Constraint(" + str(int(df_inj[sheet][well])) + " WATER_INJECTION_RATE)]"
            ws.range(rowcount + 17, 1).value = "     " + "HistoricalControlModes=UNSET"
            ws.range(rowcount + 18, 1).value = "     " + "HistoricalData=UNSET"
            ws.range(rowcount + 19, 1).value = "}"

            rowcount += 7
```

## FM FORECAST PROD & INJ (TXT MODE)

```python
import pandas as pd

df_prod = pd.read_excel("D:\GG\MINAS\SENSITIVITY LIQUID & INJECTION MFFM - PHR REQUEST AMEL\Liquid and Injection HM149_Sensitivity.xlsx", 
sheet_name = "Liquid Sensitivity", skiprows=1)
df_inj = pd.read_excel("D:\GG\MINAS\SENSITIVITY LIQUID & INJECTION MFFM - PHR REQUEST AMEL\Liquid and Injection HM149_Sensitivity.xlsx", 
sheet_name = "Injection Sensitivity", skiprows=1)

for prod in df_prod.columns[1:]:

    for index in df_prod.index:

        with open("D:\GG\MINAS\SENSITIVITY LIQUID & INJECTION MFFM - PHR REQUEST AMEL\Coding GG\\" + "Sensitivity_Rev " + str(prod) + ".txt", "a") as file:

            file.write("Well " + "\"" + str(df_prod['Wells'][index]) + "\"" + "   {\n")
            file.write("\tType=PRODUCER\n")
            
            if int(df_prod[prod][index]) != 0:
                file.write("\tStatus=OPEN\n")
            else:
                file.write("\tStatus=ALL_COMPLETIONS_SHUTIN\n")

            file.write("\tConstraints=[Constraint(" + str(int(df_prod[prod][index])) + " LIQUID_PRODUCTION_RATE)]\n")
            file.write("\tHistoricalControlModes=UNSET\n")
            file.write("\tHistoricalData=UNSET\n")
            file.write("}\n")

for inj in df_inj.columns[1:]:

    for index in df_inj.index:

        with open("D:\GG\MINAS\SENSITIVITY LIQUID & INJECTION MFFM - PHR REQUEST AMEL\Coding GG\\" + "Sensitivity_Rev " + str(inj) + ".txt", "a") as file:

            file.write("Well " + "\"" + str(df_inj['Wells'][index]) + "\"" + "   {\n")
            file.write("\tType=INJECTOR\n")
            
            if int(df_inj[inj][index]) != 0:
                file.write("\tStatus=OPEN\n")
            else:
                file.write("\tStatus=ALL_COMPLETIONS_SHUTIN\n")

            file.write("\tConstraints=[Constraint(" + str(int(df_inj[inj][index])) + " WATER_INJECTION_RATE)]\n")
            file.write("\tHistoricalControlModes=UNSET\n")
            file.write("\tHistoricalData=UNSET\n")
            file.write("}\n")
```
